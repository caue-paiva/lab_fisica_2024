```python
import numpy as np

# biblioteca de grficos
import matplotlib.pyplot as plt
from scipy.stats import norm
%matplotlib inline

"""
Cdigo baseado no repositrio https://github.com/krissiazawadzki/Lab_Fisica_Geral_2024/blob/main/pratica_1/pratica_1.ipynb
porm modificado para melhor servir aos propsitos dos experimentos da prtica 4 do lab de fsica.

Mudanas:

1) Dados inteiros com intervalos de 1 grau form considerados, levando a um processo de criao dos bins dos grfico
um pouco diferente

2) os intervalos tambm modificaram um pouco o clculo da varivel x_fit

3) Labels do grfico for modificada de acordo com o experimento

"""


TIPO_SUPERFICIE:str = "lisa" #constante para o tipo de superfcie, usada ao carregar dados e fazer as labels do grfico

# lendo os dados medidos
x_medida = np.round(np.loadtxt(f'dados_super_{TIPO_SUPERFICIE}.txt'), decimals=0)

# contando o nmero de dados
N = x_medida.shape[0]

# mdia dos dados medidos
x_medio = x_medida.mean()

# desvio padro dos dados medidos
dx_std = x_medida.std()

# mximo e mnimo das medidas
xmax = x_medida.max()
xmin = x_medida.min()
Dx = xmax - xmin

INTERVALO:int = 1 # os dados vo ter intervalos de 1 grau
bins = np.arange(xmin - 0.5, xmax + 1, INTERVALO)   #as bins vo ser os valores  entre um pouco abaixo do min e um pouco acima do mximo em intervalos de 1

x_fit = np.linspace(xmin-INTERVALO, xmax+INTERVALO, 100) #fit da gaussiana 
gauss_fit = norm.pdf(x_fit, x_medio, dx_std)

# Plotting
fig, ax = plt.subplots(figsize=(9, 6))
ax.hist(x_medida, bins=bins, color='b', width=0.8, alpha=0.5, density=True, label='dados')
ax.plot(x_fit, gauss_fit, c='r', label='gaussiana')


ax.set_xticks(np.arange(xmin, xmax+1))  
ax.set_xlabel('( ngulo medido $\pm$ 1) ')
ax.set_ylabel('porcentagem')
ax.set_title(f'medidas do ngulo crtico do objeto com superfcie {TIPO_SUPERFICIE}')

leg=ax.legend() 


```
