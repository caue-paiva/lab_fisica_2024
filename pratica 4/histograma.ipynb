```python
import numpy as np

# biblioteca de gráficos
import matplotlib.pyplot as plt
from scipy.stats import norm
%matplotlib inline

"""
Código baseado no repositório https://github.com/krissiazawadzki/Lab_Fisica_Geral_2024/blob/main/pratica_1/pratica_1.ipynb
porém modificado para melhor servir aos propósitos dos experimentos da prática 4 do lab de física.

Mudanças:

1) Dados inteiros com intervalos de 1 grau form considerados, levando a um processo de criação dos bins dos gráfico
um pouco diferente

2) os intervalos também modificaram um pouco o cálculo da variável x_fit

3) Labels do gráfico for modificada de acordo com o experimento

"""


TIPO_SUPERFICIE:str = "lisa" #constante para o tipo de superfície, usada ao carregar dados e fazer as labels do gráfico

# lendo os dados medidos
x_medida = np.round(np.loadtxt(f'dados_super_{TIPO_SUPERFICIE}.txt'), decimals=0)

# contando o número de dados
N = x_medida.shape[0]

# média dos dados medidos
x_medio = x_medida.mean()

# desvio padrão dos dados medidos
dx_std = x_medida.std()

# máximo e mínimo das medidas
xmax = x_medida.max()
xmin = x_medida.min()
Dx = xmax - xmin

INTERVALO:int = 1 # os dados vão ter intervalos de 1 grau
bins = np.arange(xmin - 0.5, xmax + 1, INTERVALO)   #as bins vão ser os valores  entre um pouco abaixo do min e um pouco acima do máximo em intervalos de 1

x_fit = np.linspace(xmin-INTERVALO, xmax+INTERVALO, 100) #fit da gaussiana 
gauss_fit = norm.pdf(x_fit, x_medio, dx_std)

# Plotting
fig, ax = plt.subplots(figsize=(9, 6))
ax.hist(x_medida, bins=bins, color='b', width=0.8, alpha=0.5, density=True, label='dados')
ax.plot(x_fit, gauss_fit, c='r', label='gaussiana')


ax.set_xticks(np.arange(xmin, xmax+1))  
ax.set_xlabel('( ângulo medido $\pm$ 1) °')
ax.set_ylabel('porcentagem')
ax.set_title(f'medidas do ângulo crítico do objeto com superfície {TIPO_SUPERFICIE}')

leg=ax.legend() 


```
